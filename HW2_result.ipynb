Exploratory Data Analysis Report: Understanding Thrombosis and Autoimmune Diseases

Thrombosis is a serious medical condition caused by the formation of blood clots within a blood vessel, leading to potentially fatal complications. Autoimmune diseases such as SLE and SJS are known to increase the risk of thrombosis. In this report, we explore a dataset provided by the authors that contains information on patients with autoimmune diseases and their thrombosis status.

Summary of Insights

Through exploratory data analysis, we have gained several key insights into the relationship between autoimmune diseases and thrombosis.

Firstly, we found that patients with thrombosis tend to have higher levels of anti-nucleus antibodies (ANA), suggesting that ANA levels could be used as a potential biomarker for predicting thrombosis. Additionally, we found that the distribution of activated partial prothrombin time (APTT) can also help to differentiate between patients with different thrombosis levels. When APTT values are greater than 35, patients are more likely to be classified as thrombosis level 1 or higher, and when APTT values are between 80 and 110, patients are more likely to be classified as thrombosis level 0.

Secondly, we found that gender plays a significant role in the distribution of certain antibodies in patients with SLE. Specifically, we found that acl_igg, acl_igm, and acl_iga levels are much lower in men than in women, while ana levels are higher in men. This suggests that gender should be considered when identifying patients with SLE.

Finally, we found that the joint distribution of Prothrombin time and APTT can be used to distinguish between patients with thrombosis level 0 and thrombosis level 2, as these two groups have non-overlapping parts in their joint distribution.
